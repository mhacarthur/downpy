```python
import os
import glob
import numpy as np
import pandas as pd

import geopandas as gpd

import cartopy.crs as ccrs
import cartopy.feature as cf
import cartopy.feature as cfeature

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
dir_base = os.path.join('/','media','arturo','T9','Data','Italy','stations','Italy_raingauges')
```

```python
dir_input = os.path.join(dir_base, 'METADATA', 'METADATA_INTENSE_NOQC.csv')
METADATA = pd.read_csv(dir_input)
METADATA.head(2)
```

```python
directories = []
for nn in range(len(METADATA)):
    file_ = METADATA['File'].iloc[nn]
    dir_ = file_.split('_')[0]
    directories.append(dir_)
METADATA['ISO'] = directories
directories = list(set(directories))
directories = sorted(directories)
```

```python
for nn in range(len(directories)):
    print(f'Directory: {directories[nn]}')
    # dir_tmp = os.path.join(dir_base,'DATA_1dy',directories[nn])
    dir_tmp = os.path.join(dir_base,'DATA_1h',directories[nn])
    if not os.path.exists(dir_tmp):
        os.makedirs(dir_tmp)
        print(f"Directory created: {dir_tmp}")
```

```python
for nn in range(len(directories)):
    META_REGION = METADATA[METADATA['ISO'] == directories[nn]].reset_index(inplace=False)

    for ss in range(len(META_REGION)):

        print(f'Station: {META_REGION['File'][ss]}')
        data = pd.read_csv(os.path.join(dir_base,'DATA',directories[nn],META_REGION['File'][ss]))
        data['Datetime_h'] = pd.to_datetime(data['Datetime_h'])
        data.set_index('Datetime_h', inplace=True)
        
        # data = data['Prec_h'].resample('1d').sum()
        # dir_out = os.path.join(dir_base,'DATA_1dy',directories[nn],META_REGION['File'][ss])
        
        data = data['Prec_h'].resample('1h').sum()
        dir_out = os.path.join(dir_base,'DATA_1h',directories[nn],META_REGION['File'][ss])

        pre_1dy = pd.DataFrame({'Datetime':data.index, 'pre':data.values})
        pre_1dy.to_csv(dir_out, header=True, index=False)
```
