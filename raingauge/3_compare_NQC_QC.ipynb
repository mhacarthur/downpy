```python
import os
import numpy as np
import pandas as pd


import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
dir_base = os.path.join('/','media','arturo','T9','Data','Italy','stations','Italy_raingauges')
```

```python
OBS_dir = os.path.join(dir_base, 'DATA_1h', 'IT-360', 'IT-360_525_0060.csv')
OBS_data = pd.read_csv(OBS_dir)
OBS_data.head(3)
```

```python
OBS_QC_dir = os.path.join(dir_base, 'DATA_1h_QC_MF', 'data_1h', 'IT-360', 'IT-360_525_0060.csv')
OBS_QC_data = pd.read_csv(OBS_QC_dir)

OBS_QC_data['Datetime'] = pd.to_datetime(OBS_QC_data['Datetime'])

OBS_QC_data.head(3)
```

```python
OBS_QC_data["color"] = np.where(OBS_QC_data["QC"] == 0, "red", "blue")
```

```python
fig = plt.figure(figsize=(5,3),dpi=300)
gs = gridspec.GridSpec(1,1)

# ============================================================================================================================
ax1 = plt.subplot(gs[0, 0])

ax1.scatter(
    x=OBS_QC_data.index,       # eje X (puede ser datetime si tu DataFrame tiene index de tiempo)
    y=OBS_QC_data["pre"].values,  # eje Y
    c=OBS_QC_data["color"].values, # array de colores
    s=1,                        # tamao del marcador
    label="NQC"                 # etiqueta
)

ax2 = ax1.twinx()
ax2.plot(OBS_QC_data.QC, 'k', linewidth=1)

ax1.tick_params(axis='both', which='major', labelsize=7)
ax2.tick_params(axis='both', which='major', labelsize=7)
```
