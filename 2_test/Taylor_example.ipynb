```python
import numpy as np
import matplotlib.pyplot as plt

def taylor_diagram(std_ref, std_model, corrcoef):
    fig = plt.figure(figsize=(6,6))
    ax = fig.add_subplot(111, polar=True)

    # Convertir correlacin a ngulo
    theta = np.arccos(corrcoef)
    ax.plot(theta, std_model, 'o', label=f'Corr={corrcoef:.2f}')
    
    # Lnea de referencia
    ax.plot([0, 0], [0, std_ref], 'k-', label='Referencia')
    
    # Ajustes del grfico
    ax.set_theta_direction(-1)
    ax.set_theta_offset(np.pi/2)
    ax.set_thetamin(0)
    ax.set_thetamax(180)
    ax.set_rlabel_position(0)
    
    ax.set_xlabel('Desviacin estndar')
    plt.legend()
    plt.title('Diagrama de Taylor')
    plt.show()

# Ejemplo con tus datos:
OBS = np.random.rand(100)
SAT = OBS + np.random.normal(0, 0.1, 100)

taylor_diagram(np.std(OBS), np.std(SAT), np.corrcoef(OBS, SAT)[0,1])
```
