```python
dir_base = os.path.join('/','media','arturo','T9','Data','Italy','Satellite','7_RE')
```

```python
data_dir = os.path.join(dir_base,f'Ensemble_RE_mean.csv')
DATA = pd.read_csv(data_dir)
RE_mean = DATA.DOWN.values
```

```python
data_dir = os.path.join(dir_base,f'Ensemble_RE_median.csv')
DATA = pd.read_csv(data_dir)
RE_median = DATA.DOWN.values
```

```python
data_dir = os.path.join(dir_base,f'Ensemble_RE_weighted.csv')
DATA = pd.read_csv(data_dir)
RE_weight = DATA.DOWN.values
```

```python
fig = plt.figure(figsize=(5,4),dpi=200)
gs = gridspec.GridSpec(1,1)

# =========================================================================================================================
ax1 = plt.subplot(gs[0, 0])

sns.kdeplot(RE_mean,color='b',linestyle='--',cumulative=False,fill=False,multiple="stack",ax=ax1,label=f'ENSEMBLE mean: {np.round(np.nanmean(RE_mean),2)}')
sns.kdeplot(RE_median,color='r',linestyle='--',cumulative=False,fill=False,multiple="stack",ax=ax1,label=f'ENSEMBLE median: {np.round(np.nanmean(RE_median),2)}')
sns.kdeplot(RE_weight,color='g',linestyle='--',cumulative=False,fill=False,multiple="stack",ax=ax1,label=f'ENSEMBLE weight: {np.round(np.nanmean(RE_weight),2)}')

ax1.set_ylabel('Percent [%]',fontsize=7)
ax1.set_xlabel('Relative Error values',fontsize=7)
ax1.legend(fontsize=6, loc=0, ncol=1)
ax1.grid(linewidth=0.3, linestyle='--')
ax1.set_xlim(-1.5,1.5)
ax1.set_ylim(0,2.5)
ax1.tick_params(axis='both', which='major', labelsize=7)
```
