## This code concat the TRMM TMPA CONUS yearly data

```python
data_base = os.path.join('..','data','tmpa_year')
```

```python
TRMM_files = sorted([f for f in os.listdir(os.path.join(data_base)) if f.endswith('.nc')])
TRMM_len = len(TRMM_files)
print(f'Number of files: {TRMM_len}')
print(f'The first time : {TRMM_files[0]}')
print(f'The last time  : {TRMM_files[-1]}')
```

```python
yy_s = int(TRMM_files[0].split('_')[-1].split('.')[0])
yy_e = int(TRMM_files[-1].split('_')[-1].split('.')[0])
print(f'Year start: {yy_s}')
print(f'Year end  : {yy_e}')
```

```python
TRMM_file_list = []
for nn in range(TRMM_len):
    TRMM_file_list.append(os.path.join(data_base,TRMM_files[nn]))
```

```python
TRMM_xr = xr.open_mfdataset(TRMM_file_list, combine='by_coords')
```
